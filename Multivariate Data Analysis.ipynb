```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
sns.set(font_scale=1.3)

cars = pd.read_csv('Dataset.csv')
```

```python
cars
```

```python
sns.pairplot(cars);
```

```python
sns.pairplot(
    data=cars, 
    aspect=.85,
    hue='transmission');
```

```python
cars.corr()
```

```python
sns.set(font_scale=1.15)
plt.figure(figsize=(8,4))
sns.heatmap(
    cars.corr(),        
    cmap='RdBu_r', 
    annot=True, 
    vmin=-1, vmax=1);
```

```python
sns.regplot(
    x='selling_price', 
    y='km_driven', 
    data=cars)
plt.xlabel(
    'Engine size in CC')
plt.ylabel(
    'Fuel efficiency');
```

```python
sns.set(font_scale=1.3)
sns.scatterplot(
    x='selling_price', 
    y='km_driven', 
    data=cars)
plt.xlabel(
    'Engine size in CC')
plt.ylabel(
    'Fuel efficiency')
```

```python
sns.scatterplot(
    x='selling_price', 
    y='km_driven', 
    data=cars,
    palette='bright',
    hue='fuel');
```

```python
sns.relplot(
    x='selling_price', 
    y='km_driven',  
    data=cars, 
    palette='bright',
    kind='scatter', 
    hue='fuel');
```

```python
sns.relplot(
    x='year', 
    y='selling_price', 
    data=cars, 
    kind='scatter', 
    col='transmission');
```

```python
sns.relplot(
    x='year', 
    y='selling_price', 
    data=cars,
    palette='bright',
    height=3, aspect=1.3,
    kind='scatter', 
    hue='transmission',
    col='fuel',
    col_wrap=2);
```

```python
sns.lmplot(
    x="year", 
    y="selling_price", 
    data=cars,
    palette='bright',
    col="transmission", 
    hue="fuel");
```

```python
sns.lineplot(
    x="year", 
    y="selling_price",
    data=cars)
```

```python
sns.lineplot(
    x="year", 
    y="selling_price",
    data=cars,
    palette='bright',
    hue='fuel');
```

```python
sns.relplot(
    x="year", 
    y="selling_price",
    data=cars,
    color='blue', height=4,
    kind='line',
    col='transmission');
```

```python
sns.relplot(
    x="year", 
    y="selling_price", 
    data=cars,
    palette='bright', 
    height=4,
    kind='line', 
    col='transmission',
    hue="fuel");
```

```python
sns.jointplot(
    x="year", 
    y="selling_price",
    kind='kde',
    shade = True,
    data=cars);
```

```python
sns.boxplot(
   x="year", 
    y="selling_price",
    data=cars,
    palette='bright',
    hue='transmission');

plt.xticks(rotation = 90)
```

```python
g = sns.catplot(
        x='year', 
        y='selling_price', 
        data=cars,
        palette='bright',
        aspect=2,
        inner=None,
        kind='violin')
sns.stripplot(
    x='year', 
        y='selling_price',
    data=cars,
    color='k', 
    linewidth=0.2,
    edgecolor='white',
    ax=g.ax);

plt.xticks(rotation = 90)
```
